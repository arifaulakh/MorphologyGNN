```python
def swc_to_nx(filename):
    G = nx.Graph()
    with open(filename) as f:
        for line in f:
            if line.startswith('#'):
                continue
            fields = line.strip().split()
            node_id, node_type, x, y, z, radius, parent_id = fields
            G.add_node(node_id, x=float(x), y=float(y), z=float(z), radius=float(radius))
            if parent_id != '-1':
                G.add_edge(parent_id, node_id)
    return G
```

```python
G = swc_to_nx("/external/rprshnas01/netdata_kcni/stlab/Public/AIBS_patchseq_2020/mouse/morphology/download.brainlib.org+8811/biccn/zeng/pseq/morph/200526/601506507_transformed.swc")
```

```python
pos = nx.spring_layout(G)

# Extract x and y coordinates from pos dictionary
x = [pos[node][0] for node in G.nodes()]
y = [pos[node][1] for node in G.nodes()]

# Draw the graph using Matplotlib
plt.figure(figsize=(8, 8))
nx.draw(G, pos, node_size=1, node_color='blue', edge_color='gray', width=0.1)
plt.savefig('swc_to_graph_viz.png')
```
